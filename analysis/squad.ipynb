## Sales combination
### Generates combination of players to sell by excluding players, and setting a target budget. 

```python
target_budget_1 = 0
target_budget_2 = 54058560

# Define excluded players for each squad
excluded_players_squad_1 = ["Casadó", "Budimir", "Lamine Yamal", "Rui Silva"]
excluded_players_squad_2 = ["Coman", "Neuer", "Olise", "Kleindienst", "Smith", "Schmid", "Claude-Maurice", "Adeyemi"]

def find_closest_combinations(squad, target, max_comb_size=15, excluded_names=None):
    # Exclude players by name if provided
    if excluded_names:
        squad = squad[~squad['name'].isin(excluded_names)]
    
    players = squad[["name", "marketValue", "playerId"]].to_dict(orient="records")
    closest_combinations = []
    
    # Generate combinations and filter those that meet or exceed the target
    for r in range(1, max_comb_size + 1):  # Loop to try combinations of size up to max_comb_size
        for combo in combinations(players, r):
            total_market_value = sum(player['marketValue'] for player in combo)
            
            # Only keep combinations where the total market value is at least the target
            if total_market_value >= target:
                difference = total_market_value - target
                closest_combinations.append((combo, total_market_value, difference))
    
    # Sort and get the top 5 closest combinations above the target
    closest_combinations = sorted(closest_combinations, key=lambda x: x[2])[:5]
    return closest_combinations

# Set the max_comb_size to control the number of players per combination
max_comb_size = 6  # Adjust this as needed

# Get the top 5 closest combinations for each squad
closest_combinations_squad_1 = find_closest_combinations(squad_1, target_budget_1, max_comb_size=max_comb_size, excluded_names=excluded_players_squad_1)
closest_combinations_squad_2 = find_closest_combinations(squad_2, target_budget_2, max_comb_size=max_comb_size, excluded_names=excluded_players_squad_2)

# Display results for squad 1
print(f"Top combinations for Squad 1 (League ID: {league_id_1}):")
for i, (combo, total_value, diff) in enumerate(closest_combinations_squad_1, 1):
    print(f"\nCombination {i} (Total Market Value: {total_value}, Difference: {diff}):")
    for player in combo:
        print(f"- {player['name']} (ID: {player['playerId']}, Market Value: {player['marketValue']})")

# Display results for squad 2
print(f"\nTop combinations for Squad 2 (League ID: {league_id_2}):")
for i, (combo, total_value, diff) in enumerate(closest_combinations_squad_2, 1):
    print(f"\nCombination {i} (Total Market Value: {total_value}, Difference: {diff}):")
    for player in combo:
        print(f"- {player['name']} (ID: {player['playerId']}, Market Value: {player['marketValue']})")
```

## Sum of players marketvalue
### Adds marketvalues of players in list.

```python
def calculate_total_market_value(squad, player_names):
    # Filter squad by player names
    selected_players = squad[squad['name'].isin(player_names)]
    total_market_value = selected_players['marketValue'].sum()
    return total_market_value

# Define player names for each league
player_names_league_1 = ["Aguerd", "Vivian", "Olasagasti", "Boyomo", "Özkaçar","Fran Pérez" ]  # Replace with your list
player_names_league_2 = ["Nsoki", "Bischof", "Gerhardt", "Scienza", "Adeyemi"]            # Replace with your list

# Calculate total market value for the specified players in each squad
total_market_value_league_1 = calculate_total_market_value(squad_1, player_names_league_1)
total_market_value_league_2 = calculate_total_market_value(squad_2, player_names_league_2)

# Display the results
print(f"Total Market Value for specified players in Squad 1 (League ID: {league_id_1}): {total_market_value_league_1}")
print(f"Total Market Value for specified players in Squad 2 (League ID: {league_id_2}): {total_market_value_league_2}")
```
